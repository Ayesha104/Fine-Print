```
#extract text from html files and get their lables from a separate csv file
import os
labelsPath = "/content/drive/My Drive/Colab Notebooks/FYP/threshold-0.75-overlap-similarity/"
policyPath="/content/drive/My Drive/Colab Notebooks/FYP/sanitized_policies/"
policies = os.listdir("/content/drive/My Drive/Colab Notebooks/FYP/sanitized_policies/")
for policy_labels_filepath in os.listdir("/content/drive/My Drive/Colab Notebooks/FYP/threshold-0.75-overlap-similarity/"):
  policy_filepath = [p_filepath for p_filepath in policies if policy_labels_filepath[0:4] in p_filepath][0]

  print("policy_labels_filepath: ", policy_labels_filepath)
  annotation_data = pd.read_csv(os.path.join(labelsPath, policy_labels_filepath), header = None)
  annotation_data = annotation_data.drop([0,1,2,3,6,7,8], axis=1)
  annotation_data = annotation_data.rename(columns={4: "segment_id", 5: "label"}) #keep only relevant columns

  print("policy_filepath: ", policy_filepath)
  print("----")
  HtmlFile = open(os.path.join(policyPath, policy_filepath), 'r', encoding='utf-8')
  html_ = HtmlFile.read() 
  soup = BeautifulSoup(html_, 'html.parser')
  policy = soup.text
  segments = policy.split("|||") #split on special character already added in dataset

  for seg_id in range(0, len(segments)):
    eachSegment = annotation_data[annotation_data["segment_id"] == seg_id]
    labels = []
    for seg_labels in eachSegment["label"]:
      labels.append(practice_dict[seg_labels]) #get all the renamed labels
    for df in practice_types: #one segment can have multiple labels
      if df in labels:
        data[df].loc[len(data[df])] = [seg_id, segments[seg_id], 1]
      else:
        data[df].loc[len(data[df])] = [seg_id, segments[seg_id], 0]

```
